```python
df = df.drop_duplicates()


df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not available')
df
```

```python
df = df.dropna(subset=['title'])
```

```python
df['title'] = df['title'].str.strip().str.lower()
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
print(df['date_added'].isna().sum())  # Counts bad/missing dates
```

```python
df['genres'] = df['listed_in'].str.split(', ')
df['countries'] = df['country'].str.split(', ')
```

```python
# Total titles
total_titles = df.shape[0]
print(f"Total titles: {total_titles}")

# Top genres
genres = df['genres'].explode()
top_genres = genres.value_counts().head(10)
print("Top 10 Genres:\n", top_genres)

# Top countries
countries = df['countries'].explode()
top_countries = countries.value_counts().head(10)
print("Top 10 Countries:\n", top_countries)

# Titles by release year
release_year_counts = df['release_year'].value_counts().sort_index()
print("Titles by Release Year:\n", release_year_counts)

# Titles by rating
rating_counts = df['rating'].value_counts()
print("Titles by Rating:\n", rating_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots appear inline in the notebook
%matplotlib inline

# Top Genres Bar Chart
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.index, y=top_genres.values,hue=top_countries.index,
            palette='magma', legend=False)
plt.xticks(rotation=45)
plt.title('Top 10 Netflix Genres')
plt.ylabel('Number of Titles')
plt.xlabel('Genre')
plt.show()

# Top Countries Bar Chart
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values, hue=top_countries.index,
            palette='magma', legend=False)

plt.xticks(rotation=90)
plt.title('Top 10 Countries by Netflix Titles')
plt.ylabel('Number of Titles')
plt.xlabel('Country')
plt.show()

# Titles by Release Year Line Chart
plt.figure(figsize=(12,6))
sns.lineplot(x=release_year_counts.index, y=release_year_counts.values, marker='o')
plt.title('Netflix Titles Released Per Year')
plt.ylabel('Number of Titles')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.show()

# Titles by Rating Pie Chart
plt.figure(figsize=(8,8))
rating_counts.plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Netflix Titles by Rating')
plt.ylabel('')
plt.show()
```
